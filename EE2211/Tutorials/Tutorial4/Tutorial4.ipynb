```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Question 2: even-determined
X = np.array([[1,1],[3,4]])
y = np.array([0,1])

w = np.linalg.inv(X)@y
print(w)
```

```python
#Question 3: over-determined

X = np.array([[1,2],[2,4],[1,-1]])
y = np.array([0,0.1,1])

w = np.linalg.inv(X.T@X)@X.T@y
print(w)
```

```python
#Question 4: under-determined system

X = np.array([[1,0,1,0],[1,-1,1,-1],[1,1,0,0]])
y = np.array([1,0,1])

w = X.T@np.linalg.inv(X@X.T)@y
print(w)
```
